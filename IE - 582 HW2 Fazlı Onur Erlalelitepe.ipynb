```r
#install.packages("imager")
library(imager)
cisem<-load.image("C:\\Users\\Linktera\\Desktop\\hw2\\hw2.jpg")
plot(cisem)
cisem_noise<-cisem
cisem_noise[,,,1]<-cisem[,,,1]+matrix(runif(256*256,min=min(cisem[,,,1]),max=max(cisem[,,,1])), ncol=256)
cisem_noise[,,,2]<-cisem[,,,2]+matrix(runif(256*256,min=min(cisem[,,,2]),max=max(cisem[,,,2])), ncol=256)
cisem_noise[,,,3]<-cisem[,,,3]+matrix(runif(256*256,min=min(cisem[,,,3]),max=max(cisem[,,,3])), ncol=256)


plot(cisem_noise)
par(mfrow=c(1,3))
image(cisem_noise[,,,1],ylim=c(1,0))#red
image(cisem_noise[,,,2],ylim=c(1,0))#green
image(cisem_noise[,,,3],ylim=c(1,0))#blue

```

```r
gray_cisem<-grayscale(cisem_noise)
a<-matrix(1:(232*232*625),ncol=625)
index=0
for (i in 1:232){
    for (j in 1:232){
        index=index+1
        a[index,]<-as.vector(gray_cisem[i:(24+i),j:(24+j)])
    }
}   

plot(gray_cisem)
```

```r
pca_a<-princomp(a)
plot(pca_a)
summary(pca_a)
# Again we have too many variables to cover for PCA and this creates a difficulty for covering the proportion of variance with 
# using a small number of components. Still, we can see that we can cover more than half of the variance with only 1 component.
#Also image of first component shows most of the details of the photo 
image(t(matrix(pca_a$scores[,1],ncol=232)),col=gray.colors(100, start = 0.3, end = 0.9, gamma = 2.2, alpha = NULL),ylim=c(1,0))
image(t(matrix(pca_a$scores[,2],ncol=232)),col=gray.colors(100, start = 0.3, end = 0.9, gamma = 2.2, alpha = NULL),ylim=c(1,0))
image(t(matrix(pca_a$scores[,3],ncol=232)),col=gray.colors(100, start = 0.3, end = 0.9, gamma = 2.2, alpha = NULL),ylim=c(1,0))

image(matrix(pca_a$loadings[,1],ncol=25),col=gray.colors(100, start = 0.3, end = 0.9, gamma = 2.2))
image(matrix(pca_a$loadings[,2],ncol=25),col=gray.colors(100, start = 0.3, end = 0.9, gamma = 2.2))
image(matrix(pca_a$loadings[,3],ncol=25),col=gray.colors(100, start = 0.3, end = 0.9, gamma = 2.2))
#from eigenvectors we can see some patterns but not a meaningful graph
```
